## 由字典建立資料框

```python
import pandas as pd
```

```python
#建立一個字典資料
data = {
    'weekday':['sun', 'sun', 'mon', 'mon'],
    'city':['Austin', 'Dallas', 'Austin', 'Dallas'],
    'visitors':[139, 237, 326, 456],
    'signups':[7, 12, 3, 5]
}
```

```python
users = pd.DataFrame(data) #利用DataFrame函數轉換data(字典)成資料框
```

```python
print(users)
```

```python
import pandas as pd
cities = ['Austin', 'Dallas', 'Austin', 'Dallas']
signups = [7, 12, 3, 5]
visitors = [139, 237, 326, 456]
weekdays = ['sun', 'sun', 'mon', 'mon']
list_labels = ['city', 'signups', 'visitors', 'weekday']
list_cols = [cities, signups, visitors, weekdays]
zipped = zip(list_labels, list_cols)
```

```python
print(zipped)
```

```python
data1 = dict(zipped)
print(data1)
```

```python
users1 = pd.DataFrame(data1)
print(users1)
```

## Importing and exporting data

### 太陽黑子觀察數數

```python
import pandas as pd
file = 'data/sunspots.csv'
sunspots = pd.read_csv(file)
sunspots.info()
```

```python
sunspots.iloc[10:20, :]
```

```python
sunspots = pd.read_csv(file,  header = None) #header
sunspots.iloc[10:20, :]
```

```python
col_names = ['year', 'month', 'day', 'frac_date', 'sunspots', 'definite']
sunspots = pd.read_csv(file,  header=None, names=col_names)
sunspots.iloc[10:20, :]
```

```python
sunspots = pd.read_csv(file,  header=None, names=col_names, na_values=' -1')
sunspots.iloc[10:20, :]
```

```python
sunspots = pd.read_csv(file,  header=None, names=col_names, na_values={'sunspots':[' -1']})
sunspots.iloc[10:20, :]
```

```python
sunspots = pd.read_csv(file,  header=None, names=col_names, na_values=' -1', parse_dates = [[0, 1, 2]])
sunspots.iloc[10:20, :]
```

```python
sunspots.info()
```

```python
sunspots.index = sunspots['year_month_day']
sunspots.index.name ='date'
sunspots.head(10)
```

### 刪除多餘欄位

```python
cols = ['sunspots','definite']
sunspots = sunspots[cols]
sunspots.iloc[10:20, :]
```
